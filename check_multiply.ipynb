```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
sizes = [100, 300, 500, 700, 1000]

for size in sizes:
  a = np.loadtxt(f'matrix_a_{size}.txt')
  b = np.loadtxt(f'matrix_b_{size}.txt')
  res = np.loadtxt(f'matrix_result_{size}.txt', skiprows=2)
  res_python = np.dot(a, b)

  if np.array_equal(res, res_python):
    print(f'Матрицы размера {size} перемножены правильно.')
  else:
    print(f'Матрицы размера {size} перемножены неправильно.')
```

```python
file = open('average_times.txt', 'r')
file = file.read()
sizes_and_times = re.findall(r"[-+]?\d*\.\d+|\d+", file)

sizes = sizes_and_times[0:len(sizes_and_times):2]
times = sizes_and_times[1:len(sizes_and_times):2]

plt.plot(sizes, times, marker='o', linestyle='-')
plt.title('Среднее время')
plt.xlabel('Размеры матриц')
plt.ylabel('Время (мс)')
plt.show()
```
